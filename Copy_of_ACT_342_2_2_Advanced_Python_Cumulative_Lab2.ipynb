Define and call a function that returns all of the models of starships.

```python
def models_starships():
  no_of_models=0
  model_name = []
  url_starships = "https://swapi.dev/api/starships"
  while url_starships is not None:
    response = requests.get(url_starships)
    data = response.json()
    for i in data['results']:
      model_name.append(i['model'])
      no_of_models=no_of_models+1
    url_starships = data['next']
  print("No of Models are:", no_of_models)
  return model_name

pp(models_starships())
```

Define and call a function that finds the ship that can carry the most cargo.

```python
def ship_carry_most_cargo():
  cargos_list = []
  ship_most_cargo = 0
  most_cargo = 0
  url_starships = "https://swapi.dev/api/starships"
  while url_starships is not None:
    response = requests.get(url_starships)
    data = response.json()
    for i in data['results']:
      if i['cargo_capacity'].isnumeric():
        cargos_list.append(float(i['cargo_capacity']))
    url_starships = data['next']
  most_cargo = max(cargos_list)
  print(most_cargo)

  url_starships = "https://swapi.dev/api/starships"
  while url_starships is not None:
    response = requests.get(url_starships)
    data = response.json()
    for i in data['results']:
      if i['cargo_capacity'].isnumeric():
        if(most_cargo == float(i['cargo_capacity'])):
          ship_most_cargo = i['name']
    url_starships = data['next']
  return ship_most_cargo

ship_carry_cargo = ship_carry_most_cargo()
print("The ship which carries the most cargo is", ship_carry_cargo)
```

```python
def crew_passengers(starship_name):
  num_of_crew = 0
  num_of_passengers = 0
  url_starships = "https://swapi.dev/api/starships"
  while url_starships is not None:
    response = requests.get(url_starships)
    data = response.json()
    for i in data['results']:
      if i['name'] == starship_name:
        num_of_crew = i['crew']
        num_of_passengers = i['passengers']
    url_starships = data['next']
  return num_of_crew, num_of_passengers

no_of_crew, no_of_passengers = crew_passengers('Belbullab-22 starfighter')
print("No of crew is", no_of_crew)
print("No of passengers is", no_of_passengers)
```
