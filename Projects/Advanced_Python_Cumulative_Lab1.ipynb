Define a function that returns a list of all the charachter names.

Call the function.

```python
def char_names():
  no_of_character=0
  char_name = []
  url_people = "https://swapi.dev/api/people"
  no_of_character = 0
  while url_people is not None:
    response = requests.get(url_people)
    data = response.json()
    for i in data['results']:
      char_name.append(i['name'])
      no_of_character=no_of_character+1
    url_people = data['next']
  print("No of Characters are:", no_of_character)
  return char_name


pp(char_names())
```

Define a function that counts the total number of male charachters and the total number of female charachters. Return both numbers.

Call the function.
<br></br>
Considerations: Do all charachters have the gender male or female?

```python
def find_gender():
  url_people = "https://swapi.dev/api/people"
  male_characters = 0
  female_characters = 0
  others = 0
  while url_people is not None:
    response = requests.get(url_people)
    data = response.json()
    for character in data['results']:
      if(character['gender']=="male"):
        male_characters=male_characters+1
      elif(character['gender']=="female"):
        female_characters=female_characters+1
      else:
        others = others+1
      url_people = data['next']
  return male_characters, female_characters, others


male_char, female_char, other_char = find_gender()
print("No of Male characters are", male_char)
print("No of Female characters are", female_char)
print("No of other characters are", other_char)



```

Define a function that returns a list of all the charachters that appear in more than 'x' number of films. Where 'x' is an integer that is passed as an argument into the function.

Call the function with at least two test cases.
<br></br>
Considerations: Is 'more' than inclusive of the number passed?

```python
def find_artist_count(no_of_films):
  url_people = "https://swapi.dev/api/people"
  characters = 0
  char_name = []
  while url_people is not None:
    response = requests.get(url_people)
    data = response.json()
    for character in data['results']:
      #print(len(character['films']))
      if(len(character['films']) >= no_of_films):
        char_name.append(character['name'])
        characters=characters+1
      url_people = data['next']
  return no_of_films, char_name, characters

films_count, artist_name, num_of_artisits = find_artist_count(3)
pp(artist_name)
print("No of artists who acted in {} or more films are {}" .format(films_count, num_of_artisits))
```

Define a function that returns a list of all the charachters that start with a specific letter. Where the letter is pass as a string as an argument into the function.

Call the function with an uppercase letter.
Call the function with the same charachter as a lowercase letter.
<br></br>
Considerations: How do you handle the type of charachter being passed? There are many different ways to make this work. Be Creative.

```python
def find_artist_name(first_letter):
  url_people = "https://swapi.dev/api/people"
  characters = 0
  char_name = []
  while url_people is not None:
    response = requests.get(url_people)
    data = response.json()
    for character in data['results']:
      if(character['name'][0]) == first_letter.upper():
        char_name.append(character['name'])
        characters=characters+1
      url_people = data['next']
  return first_letter, char_name, characters

starting_letter, artist_name, num_of_artisits = find_artist_name("B")
pp(artist_name)
print("No of artists name starts with the letter {} are {}" .format(starting_letter, num_of_artisits))
```
